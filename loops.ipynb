Loop while

```python
n = int(input('qtde de números: '))
cont = 1 
s = 0
while cont < n:
    s = s + cont
    cont = cont + 1
    print(s)

```

```python
#critério de parada não sendo o contador
cont = 1
x = 0
while x < 20:
    x = cont * 5
    cont = cont + 1
    print(x)
```

```python
#unindo com if

nomes = ['mario','jose','mario','maria','carlos','ana','mario']
n = len(nomes)
cont = 1
while cont <= n-1:
    if nomes[cont] != 'mario':
        print(nomes[cont])
    cont = cont + 1
```

```python
#usando break para parar a soma quando o primeiro resultado par aparecer
n = int(input('qtde de números: '))
cont = 1
s = 0
while cont <= n:
    s = s + cont
    if s % 2 == 0 : break
    cont = cont + 1
    print(s)
```

For


```python
#diferenemene do while, não executa somente enquanto uma condição é True. Executa de qualquer maneira
n = int(input('total de repetições = '))
for i in range(0,n):
    print('bovespa')
```

```python
lista_nomes = ['um','dois','tres','quatro','cinco']

for i in range(0, len(lista_nomes)):
    print(lista_nomes[i])
```

```python
lista_nomes = ['um','dois','tres','quatro','cinco']
nova=[]
for i in range(0,len(lista_nomes)):
    nova.append(lista_nomes[i])
    print(nova)
```

```python
#fatorial 
n = int(input('n = '))
fat = 1
for i in range(1, n+1):
    fat = fat * i
print('fatorial = ',fat)
```
