```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,20)
y = 3*x-3
```

```python
plt.plot(x,y)#plotar grafico
plt.title('função linear')#titulo do grafico
plt.xlabel('eixo x')#nome do eixo x
plt.ylabel('eixo y')#nome do eixo y
plt.show()
#plt.savefig('grafico.png')#salvar grafico

```
