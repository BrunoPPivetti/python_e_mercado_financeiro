7 - Observar a seguinte lista: lista = [2,2,3,3,3,-1,-1,-2,0,0,0,2,4,5,1,2,2,0,0,0,2,1,5,5,7,6,5,0,0]. Encontre as seguintes estatisticas:
- soma
- maximo
- minimo
- media
- mediana
- moda
- desvio-padrao populacional
- desvio-padrao amostral
- quantidade de vezes que aparece o numero 0
- quantidade de vezes que aparece o numero 5
- ordenar a lista em ordem crescente
- ordenar a lista em ordem decrescente

```python
lista = [2,2,3,3,3,-1,-1,-2,0,0,0,2,4,5,1,2,2,0,0,0,2,1,5,5,7,6,5,0,0]

print(np.sum(lista))
print(np.max(lista)) 
print(np.min(lista)) 
print(np.mean(lista))
print(np.median(lista))
print(st.mode(lista))
print(st.stdev(lista))
print(st.pstdev(lista))
print(f'o numero zero aparece: {lista.count(0)} vezes')
print(f'o numero zero aparece: {lista.count(5)} vezes')
print(sorted(lista))
print(sorted(lista, reverse=True))
```

8 - Através da planilha exer8:
- importe os dados do excel e transforme a coluna A em uma variável e represente a Vale e a coluna B em outra variável que represente a coluna B
- transforme as variáveis em vetores usando a biblioteca numpy
- fazer os dois gráficos dos preços da vale e da gerdau usando subplot. colocar a vale na parte superior da figura e a gerdau na inferior
- calcular os retornos das duas empreas e plotas os quatro graficos no formato de matriz 2x2

Era para ser através da lib xlrd, mas atualizou e não é mais possível. Por isso, vou usar pandas. (muito mais simples também)


```python
df = pd.read_excel('exer8.xlsx')
print('A) transformar as colunas A e B em colunas da Vale e Gerdau')
vale = df['A']
gerdau = df['B']
print(vale.head())
print(gerdau.head())

print('b) transformar as variáveis em vetores')
print('como usei o pandas, já são.')

print('C) fazer dois gráficos dos preços da Vale e da Gerdau usando subplot. Vale na parte superior da figura e gerdau na inferior')
plt.figure(figsize=(10,5))
plt.subplot(2,1,1)
plt.plot(vale)
plt.title('Value')
plt.subplot(2,1,2)
plt.plot(gerdau)
plt.title('Gerdau')
plt.show()

print('D) calcular retornos e plotar em um gráfico em matriz 2x2 junto com o preço das ações.')
retornos_vale = vale.pct_change()
retornos_gerdau = gerdau.pct_change()
plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.plot(vale)
plt.title('Vale')
plt.subplot(3,2,2)
plt.plot(gerdau)
plt.title('Gerdau')
plt.subplot(3,2,3)
plt.plot(retornos_vale)
plt.title('Retornos Vale')
plt.subplot(3,2,4)
plt.plot(retornos_gerdau)
plt.title('Retornos Gerdau')

plt.show()
```
