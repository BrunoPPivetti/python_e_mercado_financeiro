```python
#retorno esperado = somatório de todos os retornos esperados * probabilidade
r1 = float(input('Digite o retorno esperado da primeira ação: '))
r2 = float(input('Digite o retorno esperado da segunda ação: '))
r3 = float(input('Digite o retorno esperado da terceira ação: '))

p1 = float(input('Digite a probabilidade da primeira ação: '))
p2 = float(input('Digite a probabilidade da segunda ação: '))
p3 = float(input('Digite a probabilidade da terceira ação: '))

retorno_esperado = (r1 * p1) + (r2 * p2) + (r3 * p3)
print(retorno_esperado)
```
