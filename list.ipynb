revisão basica de listas

```python
mercado = ['ações','opções','futuro','dolar','ouro','criptomoedas']
mercado
```

```python
#slice 
mercado[0:3]
#o ultimo número não é incluído. O python considera sempre o primeiro até N-1
```

```python
#"in" -> verificar se o valor está presente na lista (boolean)
print('ações' in mercado)
print('b3' in mercado)
fut = 'futuro' in mercado
print('futuro está na lista?',fut)
```

```python
#substituindo valor
mercado[2] = 'futuros de commodities'
print(mercado)

#alterando mais de um valor. use-se o fatiamento
mercado[0:1] = ['titulos','tesouro']
print(mercado)
```

```python
#adicionando valores
mercado.append('bdrs')
print(mercado)
```

```python
mercado.count('titulos')
```

```python
#juntando uma lista com a outra
mercado.extend(['calls','puts','ntnb'])
print(mercado)
```

```python
#colocando em ordem alfabética ou crescente
mercado.sort()
print(mercado)
#para ordem ao contrário
mercado.sort(reverse=True)
print(mercado)
#se houverem letras maiúsculas e minúsculas, ele coloca as maiusculas primeiro. Por isso, usa-se mercado.sort(key=str.casefold)
```

```python
#remover itens de uma lista
mercado.remove('ntnb')
print(mercado)
```

```python
#para inserir um elemento em uma posição específica, usa-se mercado.insert(posição, elemento). O append adiciona o elemento no final da lista
mercado.insert(2,'Petr4')
print(mercado)
```

```python
#limpando a lista
mercado.clear()
print(mercado)
```

```python
ibov = ['petr4', 'itub4', 'vale3','ggbr4','axia6']
#do indice 2 ao indice 3
print(ibov[2:3])
#do indice 1 ao final
print(ibov[1:])
#do indice 0 ao indice 2, já que N-1
print(ibov[:3])
#do inicial ao ultimo pulando de 2 em 2
print(ibov[::2]) #ou poderia ser print(ibov[0:5:2])
#os ultimos 3 elementos da lista
print(ibov[-3:])
#os dois primeiros elementos da lista (bem mais fácil usar ibov[:2])
print(ibov[:-3])
print(ibov[:2])
```
