```python
l1 = int(input('lado 1: '))
l2 = int(input('lado 2: '))
l3 = int(input('lado 3: '))
if l1 == l2 == l3:
    print('triangulo equilátero')
elif l1 == l2 or l2 == l3 or l1 == l3:
    print('triangulo isosceles')
else: 
    print('triangulo escaleno')
```

O suporte de uma ação é calculado como 30% do intervalor histórico de uma ação (max - min). A resistência é calculada como o valor de 60% do intervalo histórico.
Faça um programa que o usuário forneça os valores mais baixo e mais alto através de um input. O programa pede o valor atual da ação com input e diz através de print() qual é o suporte e a resistência e se o preço está no sup ou na res ou fora dela.

```python
preco = int(input('insira o preço da ação: '))
max = int(input('insira o preço máximo da ação'))
min = int(input('insira o preço mínimo do ação: '))

res = (max - min) * 0.6
sup = (max- min) * 0.3

if preco < res and preco > sup:
    print('O preço está dentro da faixa sup-res')
else: 
    print('O preço está fora da faixa sup-res')
```

5 - Considere uma equação do segundo grau: Ax² + Bx + C = 0. Utilizando delta = B² - 4AC, Escreva um programa que diga se não há raízes, se há uma única raiz ou se há duas raízes

```python
import math

a = int(input('insira o valor de a:'))
b = int(input('insira o valor de b:'))
c = int(input('insira o valor de c:'))

delta = b*b - 4*a*c

if delta < 0:
    print('não há raiz.')
elif delta == 0:
    print('existe uma raiz')
else: 
    x1 = -b + math.sqrt(delta)/2*a
    x2 = -b - math.sqrt(delta)/2*a
    print(f'há duas raízes: x1 {x1} e x2 {x2}')
```

7 - inputar uma sequência de números naturais, o algoritmo faz o soma para saber o total de vendas e posteriormente exibe a soma dos números pares e dos números ímpares.

```python
total_vendas = int(input('digite o total de vendas: '))

sp = 0
si = 0

for i in range(total_vendas):
    x = int(input('x = '))
    if x % 2 == 0:
        sp += x
    else:
        si += x
print('soma pares:',sp, 'soma impares',si)
```

8 - O financiamento de um automóvel foi contratado respeitando a série matemática a seguir para 70 meses. Criar um algoritmo em Python para calcular a soma seguinte até o termo n que o usuáriod esejar, seguindo a lógica a seguir.
S = 70/7 + 69/14 + 68/21 + 67/28

10 - usuario entre com a variável n (quantidade de termos de uma integral) e x (limite de integração) e o programa exibe o resultado.


```python
n = int(input('n = '))
x = float(input(' x = '))
s = x 
imp = 1
fat = 1
for i in range(1,n):
    fat = fat * i
    imp = imp + 2
    s = s + (x**imp)*(-1)**i/(imp*fat)

print('soma: ', s)
```
