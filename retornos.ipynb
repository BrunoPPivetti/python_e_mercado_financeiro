```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('opçao.xlsx')
#no livro o autor usar xlrd, portanto precisa transformar as colunas em array para as operações, como usei o pandas, já são vetorizadas
```

```python
vetor = df['A']
```

```python
retorno = (vetor - vetor.shift(1))/vetor.shift(1) #por estar usando pandas 
#se fosse como no livro:
#(vetor[1:20] - vetor[0:19])/(vetor[0:19])
retorno
```

```python
t = np.arange(0,20) #cria uma sequencia de 0 a 19
plt.plot(t, retorno)
plt.title('retorno')
plt.xlabel('tempo')
plt.ylabel('retorno')
plt.grid()
plt.show()
```

```python
#calculando cenários otimista e pessimista
#formula: retorno_médio + 2*desvio_padrao/sqrt(n) ou retorno_médio - 2*desvio_padrao/sqrt(n)

media = st.mean(retorno.dropna())
desvio = st.pstdev(retorno.dropna())
otimista = media+2*desvio/math.sqrt(19) #poderia ser numpy
pessimista = media-2*desvio/math.sqrt(19)
print(f"otimista: {otimista*100:.2f}%")
print(f"pessimista: {pessimista*100:.2f}%")
```
